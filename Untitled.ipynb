```python
import pandas as pd
import matplotlib.pyplot as plt
import glob

files = glob.glob(r'C:\Users\ACC1\Desktop\Project\Big_Project\*.xlsx')

df_list = []
for file in files:
    df_list.append(pd.read_excel(file))

df = pd.concat(df_list, ignore_index=True)
df = df.drop(columns=['Mobile', 'Tower', 'Flat', 'Name'], errors='ignore')
df = df.dropna(subset=['Building', 'Customer_ID'])
df = df[df['Category'] != 'Sample']
df = df[df['Order_Total'] > 0]
df = df.dropna(subset=['Order_Total'])

bulk_index = df[df['Order_Total'] == 37502.1].index
df = df.drop(bulk_index[1:])

df['Date'] = pd.to_datetime(df['Date'])

print(df.shape)
print("Date range:", df['Date'].min(), "to", df['Date'].max())
```

```python
# Before daily delivery — up to October 31 2025
before = df[df['Date'] < '2025-11-01']
# After daily delivery — November 2025 onwards
after = df[df['Date'] >= '2025-11-01']
print("BEFORE Daily Delivery:")
print("Row :",len(before))
print("Date Range:",before['Date'].min(),"to",before['Date'].max())
print("Total Revenue:",before['Order_Total'].sum().round(2))
print("Unique Customer:",before['Customer_ID'].nunique())

print("\nAfter Daily Delivery:")
print("Row :",len(after))
print("Date Range:",after['Date'].min(),"to",after['Date'].max())
print("Total Revenue:",after['Order_Total'].sum().round(2))
print("Unique Customer:",after['Customer_ID'].nunique())
```

```python
# Monthly average comparison
before_months = before['Date'].dt.to_period('M').nunique()
after_months = after['Date'].dt.to_period('M').nunique()

before_monthly_avg_revenue = before['Order_Total'].sum() / before_months
after_monthly_avg_revenue = after['Order_Total'].sum() / after_months

before_monthly_avg_orders = len(before) / before_months
after_monthly_avg_orders = len(after) / after_months

before_monthly_avg_customers = before.groupby(before['Date'].dt.to_period('M'))['Customer_ID'].nunique().mean()
after_monthly_avg_customers = after.groupby(after['Date'].dt.to_period('M'))['Customer_ID'].nunique().mean()

print("MONTHLY AVERAGES COMPARISON")
print(f"{'Metric':<30} {'Before':>15} {'After':>15} {'Change':>10}")
print("-" * 70)
print(f"{'Avg Monthly Revenue (₹)':<30} {before_monthly_avg_revenue:>15,.0f} {after_monthly_avg_revenue:>15,.0f} {((after_monthly_avg_revenue/before_monthly_avg_revenue)-1)*100:>+9.1f}%")
print(f"{'Avg Monthly Orders':<30} {before_monthly_avg_orders:>15,.0f} {after_monthly_avg_orders:>15,.0f} {((after_monthly_avg_orders/before_monthly_avg_orders)-1)*100:>+9.1f}%")
print(f"{'Avg Monthly Customers':<30} {before_monthly_avg_customers:>15,.0f} {after_monthly_avg_customers:>15,.0f} {((after_monthly_avg_customers/before_monthly_avg_customers)-1)*100:>+9.1f}%")
```

```python
before_avg_order = before['Order_Total'].sum() / len(before)
after_avg_order = after['Order_Total'].sum() / len(after)

print(f"Average order value BEFORE: ₹{before_avg_order:.0f}")
print(f"Average order value AFTER: ₹{after_avg_order:.0f}")
print(f"Change: {((after_avg_order/before_avg_order)-1)*100:.1f}%")
```

```python
# Revenue per customer per month — before vs after
before_rev_per_customer = before_monthly_avg_revenue / before_monthly_avg_customers
after_rev_per_customer = after_monthly_avg_revenue / after_monthly_avg_customers

print(f"Monthly revenue per customer BEFORE: ₹{before_rev_per_customer:.0f}")
print(f"Monthly revenue per customer AFTER: ₹{after_rev_per_customer:.0f}")
print(f"Change: {((after_rev_per_customer/before_rev_per_customer)-1)*100:.1f}%")
```

```python
# Compare same time period last year vs this year
same_period_before = df[(df['Date'] >= '2024-11-01') & (df['Date'] < '2025-11-01')]
same_period_after = df[df['Date'] >= '2025-11-01']

# Monthly averages for same period
spb_months = same_period_before['Date'].dt.to_period('M').nunique()
spa_months = same_period_after['Date'].dt.to_period('M').nunique()

spb_revenue = same_period_before['Order_Total'].sum() / spb_months
spa_revenue = same_period_after['Order_Total'].sum() / spa_months

spb_customers = same_period_before.groupby(same_period_before['Date'].dt.to_period('M'))['Customer_ID'].nunique().mean()
spa_customers = same_period_after.groupby(same_period_after['Date'].dt.to_period('M'))['Customer_ID'].nunique().mean()

spb_rev_per_customer = spb_revenue / spb_customers
spa_rev_per_customer = spa_revenue / spa_customers

print("SAME PERIOD COMPARISON — Nov to Jun")
print(f"{'Metric':<35} {'2024-25':>12} {'2025-26':>12} {'Change':>10}")
print("-" * 70)
print(f"{'Avg Monthly Revenue (₹)':<35} {spb_revenue:>12,.0f} {spa_revenue:>12,.0f} {((spa_revenue/spb_revenue)-1)*100:>+9.1f}%")
print(f"{'Avg Monthly Customers':<35} {spb_customers:>12,.0f} {spa_customers:>12,.0f} {((spa_customers/spb_customers)-1)*100:>+9.1f}%")
print(f"{'Revenue per Customer (₹)':<35} {spb_rev_per_customer:>12,.0f} {spa_rev_per_customer:>12,.0f} {((spa_rev_per_customer/spb_rev_per_customer)-1)*100:>+9.1f}%")
```

```python
# Society level comparison — same period

before_society = same_period_before.groupby('Building')['Order_Total'].sum()

after_society = same_period_after.groupby('Building')['Order_Total'].sum()

society_comparison = pd.DataFrame({

    'Before_Nov24_Jun25': before_society,

    'After_Nov25_Jun26': after_society

}).fillna(0)

society_comparison['Change_Pct'] = ((society_comparison['After_Nov25_Jun26'] / 

    society_comparison['Before_Nov24_Jun25']) - 1) * 100

society_comparison = society_comparison[society_comparison['Before_Nov24_Jun25'] > 0]

society_comparison = society_comparison.sort_values('Change_Pct', ascending=False)

print("TOP 10 GROWING SOCIETIES:")

print(society_comparison.head(10))

print("\nTOP 10 DECLINING SOCIETIES:")

print(society_comparison.tail(10))
```

```python
# Check how many buildings look like individual addresses
total_buildings = df['Building'].nunique()
print("Total unique buildings:", total_buildings)

# Check Crescent Bay specifically
cb_before = same_period_before[same_period_before['Building'] == 'Crescent Bay']['Order_Total'].sum()
cb_after = same_period_after[same_period_after['Building'] == 'Crescent Bay']['Order_Total'].sum()

print(f"\nCreescent Bay Nov-Jun 2024-25: ₹{cb_before:,.0f}")
print(f"Crescent Bay Nov-Jun 2025-26: ₹{cb_after:,.0f}")
print(f"Change: {((cb_after/cb_before)-1)*100:.1f}%" if cb_before > 0 else "No data before")
```

```python
# All Crescent Bay customers ever
cb_all = df[df['Building'] == 'Crescent Bay'].copy()

# Last order date per customer
cb_customers = cb_all.groupby('Customer_ID').agg(
    first_order=('Date', 'min'),
    last_order=('Date', 'max'),
    total_orders=('Date', 'count'),
    total_revenue=('Order_Total', 'sum')
).reset_index()

# Days since last order
latest_date = df['Date'].max()
cb_customers['days_since'] = (latest_date - cb_customers['last_order']).dt.days

# Segment
cb_customers['status'] = cb_customers['days_since'].apply(
    lambda x: 'Active' if x <= 60 else ('At Risk' if x <= 180 else 'Lost')
)

print("Crescent Bay customer status:")
print(cb_customers['status'].value_counts())
print(f"\nTotal CB customers ever: {len(cb_customers)}")
print(f"\nLost customers detail:")
lost = cb_customers[cb_customers['status'] == 'Lost'].sort_values('last_order', ascending=False)
print(lost[['Customer_ID', 'first_order', 'last_order', 'total_orders', 'total_revenue', 'days_since']].head(20))
```
